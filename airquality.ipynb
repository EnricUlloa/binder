```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(df["Wind"], df["Solar.R"], color="blue", alpha=0.5)

plt.xlabel("Velocitat del vent (Wind)")
plt.ylabel("Radiació Solar (Solar.R)")
plt.title("Relació entre el Vent i la Radiació Solar")

plt.show()
```

Este grafico representa la relacion entre el Viento y la Radiacion Solar. Se puede ver que cuando la velocidad del viento esta por la mitad (7,5 - 15,0) es donde se centra mas radiacion solar.

```python
import matplotlib.pyplot as plt
import seaborn as sns

temp_mitjana_per_mes = df.groupby("Month")["Temp"].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=temp_mitjana_per_mes.index, y=temp_mitjana_per_mes.values, palette="Blues")

plt.xlabel("Mes")
plt.ylabel("Temperatura Mitjana")
plt.title("Temperatura Mitjana per Mes")

plt.show()
```

Como se puede observar, esta grafica representa la temperaura media por mes. Podemos apreciar que en el mes 7 y 8 han alcanzado la temperatura media de 80. Mientras que el mes 5 es el que ha registrado una temperatura media de 65.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Ozone"], y=df["Temp"], color="red", alpha=0.7)

plt.xlabel("Concentració d'Ozó (Ozone)")
plt.ylabel("Temperatura (Temp)")
plt.title("Relació entre l’Ozó i la Temperatura")

plt.show()
```

Esta grafica representa una relacion entre el ozono y la temperatura. Podemos ver que como a mas concentracion (100 - 175) de ozono hay menos temperatura.
