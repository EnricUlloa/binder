```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("militaryTanks.csv")
```

Como se puede ver con el "df.head()" muestra las 5 primeras filas.

```python
max_range = df["Range (km)"].max()
print(max_range)
```

Con esto logramos sacar el rango maximo del dataset.

```python
fila_max = df[df["Range (km)"] == df["Range (km)"].max()]
print(fila_max)
```

Con esto lo que conseguimos es mostrar que tanque tiene el mayor rango, que en este caso es el tanque M-95 Degman de la nacion Croata.

```python
fila_max = df[df["Range (km)"] == df["Range (km)"].max()]
print(fila_max[["Model Name", "Range (km)"]])
```

Esto de aqui es una forma mas precisa de saber cual es el tanque con el mayor rango. El rango significa que tanto recorrera con solo un tanque de combustible.

```python
tank_counts = df["Origin Country"].value_counts()
print(tank_counts)
```

Esto es para ver cuantos tanques tiene cada nacion en este dataset. Podemos ver que China tiene 5 tanques diferentes.

```python
import pandas as pd
import matplotlib.pyplot as plt

tank_counts = df["Origin Country"].value_counts()

tank_counts.plot(kind="bar")
plt.xlabel("País")
plt.ylabel("Cantidad de tanques")
plt.show()
```

Esto de aqui es una grafica que muestra la cantidad de tanques por pais.

```python
df["Weight (in Tons)"].plot(kind="hist")
plt.show()
```

Podemos ver que la mayoria de tanques pesan entre 40 aprox a 45 toneladas.

```python
df["Main Weapon Caliber"].value_counts().plot(kind="pie")
plt.show()
```

Aqui podemos ver que la mayoria de tanques optan por un cañon de 125 mm, en segundo lugar de 105, luego 120 y por ultimo el de 90 mm.
